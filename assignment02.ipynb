```python
import numpy as np
import matplotlib.pyplot as plt
import random
from scipy.misc import derivative
```

# Define a 2-dimentional function, $f(x)=x^2$

```python
def givenFunction(x):
    f = x ** 2
    return f
```

# Create a function for getting first derivative of any given function.

```python
def firstDerivative(x):
    return derivative(givenFunction, x, dx = 1e-4)
```

# Set the domain for the given function.

```python
x = np.arange(-15, 15, 0.1)
```

# Randomly pick 3 points from the defined domain.

```python
points = np.empty(3)
print("Three picked points: ")
for i in range(0, 3):
    points[i] = random.choice(x)
    print(points[i])
```

# Define the alfine function after taken taylor approximation.

```python
def taylorAprroximation(x, xPoint, yPoint):
    f = firstDerivative(xPoint) * (x - xPoint) + yPoint
    return f
```

# Compute the result graphs.

```python
f = givenFunction(x)
aF = np.empty((3,x.size))
for i in range(0, 3):
    aF[i] = taylorAprroximation(x, points[i], givenFunction(points[i]))
```

# Plot selected 3 points, 3 taylor approximations, and the original function.

```python
plt.figure(1)
plt.plot(x, f, 'b', label="original function")
plt.xlim(-20, 20)
for i in range(0, 3):
    plt.plot(points[i], givenFunction(points[i]), 'o', label="picked point %d" %i)
for i in range(0, 3):
    plt.plot(x, aF[i], 'r', label="taylor approximation at x = %d" % points[i])
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
